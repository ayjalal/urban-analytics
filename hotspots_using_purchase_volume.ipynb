```python
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import HeatMap
```

```python


# Step 1: Load the large dataset
df = pd.read_csv("datasets/Tabelle_Purchase_Value.csv")  # Replace with your CSV file path

# Step 2: Randomly sample 900,000 rows from the dataset
sample_size = 350000
sampled_df = df.sample(n=sample_size, random_state=42)  # Ensures reproducibility

# Step 3: Check your column names for D_CUS_DeliveryLatitude and D_CUS_DeliveryLongitude
D_CUS_DeliveryLatitude_column = 'D_CUS_DeliveryLatitude'  # Replace with your actual D_CUS_DeliveryLatitude column name
D_CUS_DeliveryLongitude_column = 'D_CUS_DeliveryLongitude'  # Replace with your actual D_CUS_DeliveryLongitude column name

# Step 4: Create a map centered around the average location
center_lat = sampled_df[D_CUS_DeliveryLatitude_column].mean()
center_lon = sampled_df[D_CUS_DeliveryLongitude_column].mean()
heatmap_map = folium.Map(location=[center_lat, center_lon], zoom_start=10)

# Step 5: Prepare data for the HeatMap
heat_data = sampled_df[[D_CUS_DeliveryLatitude_column, D_CUS_DeliveryLongitude_column]].dropna().values.tolist()

# Step 6: Add the HeatMap to the map
HeatMap(heat_data, radius=12).add_to(heatmap_map)

# Step 7: Add a legend to the map
legend_html = '''
<div style="
    position: fixed; 
    bottom: 550px; left: 390px; width: 150px; height: 100px; 
    background-color: white; z-index:1000; font-size:14px;
    border:1px solid black; padding: 10px;">
    <b>Heatmap Intensity</b><br>
    <i style="background:yellow; width:10px; height:10px; display:inline-block;"></i> Low<br>
    <i style="background:orange; width:10px; height:10px; display:inline-block;"></i> Medium<br>
    <i style="background:red; width:10px; height:10px; display:inline-block;"></i> High<br>
</div>
'''
heatmap_map.get_root().html.add_child(folium.Element(legend_html))

# Step 8: Save the map to an HTML file and display it
heatmap_map.save('heatmap16.html')
print("Heatmap saved as 'heatmap7.html'")
```

Bars density and buffer zones in Mitte

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import geopandas as gpd
from shapely.geometry import Point

# Step 1: Load the large dataset
df = pd.read_csv("datasets/Tabelle_contains_Alkohol.csv")

# Step 2: Randomly sample 250,000 rows from the dataset
sample_size = 250000
sampled_df = df.sample(n=sample_size, random_state=42)

# Step 3: Column names
D_CUS_DeliveryLatitude_column = 'D_CUS_DeliveryLatitude'
D_CUS_DeliveryLongitude_column = 'D_CUS_DeliveryLongitude'

# Step 4: Load the boundary of Mitte
mitte = gpd.read_file('mitte_district.geojson')
# Remove any index columns
mitte = mitte.loc[:, ~mitte.columns.str.contains('^index')]

# Step 5: Create a GeoDataFrame from the sampled data
geometry = [Point(xy) for xy in zip(sampled_df[D_CUS_DeliveryLongitude_column], sampled_df[D_CUS_DeliveryLatitude_column])]
sampled_gdf = gpd.GeoDataFrame(sampled_df, geometry=geometry, crs="EPSG:4326")
# Remove any index columns
sampled_gdf = sampled_gdf.loc[:, ~sampled_gdf.columns.str.contains('^index')]

# Step 6: Filter the data to include only points within Mitte
sampled_gdf_within_mitte = gpd.sjoin(sampled_gdf, mitte, predicate='within')
# Remove any index columns
sampled_gdf_within_mitte = sampled_gdf_within_mitte.loc[:, ~sampled_gdf_within_mitte.columns.str.contains('^index')]

# Step 7: Create a map centered around the average location
center_lat = sampled_gdf_within_mitte[D_CUS_DeliveryLatitude_column].mean()
center_lon = sampled_gdf_within_mitte[D_CUS_DeliveryLongitude_column].mean()
heatmap_map = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Step 8: Prepare data for the HeatMap
heat_data = sampled_gdf_within_mitte[[D_CUS_DeliveryLatitude_column, D_CUS_DeliveryLongitude_column]].dropna().values.tolist()

# Step 9: Add the HeatMap for alcohol purchases
HeatMap(heat_data, radius=12).add_to(heatmap_map)

# Step 10: Add legend
legend_html = '''
<div style="
    position: fixed; 
    bottom: 550px; left: 390px; width: 150px; height: 100px; 
    background-color: white; z-index:1000; font-size:14px;
    border:1px solid black; padding: 10px;">
    <b>Heatmap Intensity</b><br>
    <i style="background:yellow; width:10px; height:10px; display:inline-block;"></i> Low<br>
    <i style="background:orange; width:10px; height:10px; display:inline-block;"></i> Medium<br>
    <i style="background:red; width:10px; height:10px; display:inline-block;"></i> High<br>
</div>
'''
heatmap_map.get_root().html.add_child(folium.Element(legend_html))

# Step 11: Load and prepare bars data
bars = gpd.read_file('bars.geojson')
# Remove any index columns
bars = bars.loc[:, ~bars.columns.str.contains('^index')]

# Step 12: Filter bars to Mitte district
bars_in_mitte = gpd.sjoin(bars, mitte, predicate='within')
# Remove any index columns
bars_in_mitte = bars_in_mitte.loc[:, ~bars_in_mitte.columns.str.contains('^index')]

# Step 13: Create buffer zones
bars_in_mitte_buffered = bars_in_mitte.copy()
bars_in_mitte_buffered['geometry'] = bars_in_mitte_buffered.geometry.buffer(0.005)

# Step 14: Remove any index columns before final spatial join
sampled_gdf_within_mitte = sampled_gdf_within_mitte.loc[:, ~sampled_gdf_within_mitte.columns.str.contains('^index')]
bars_in_mitte_buffered = bars_in_mitte_buffered.loc[:, ~bars_in_mitte_buffered.columns.str.contains('^index')]

# Perform the spatial join
purchase_within_buffers = gpd.sjoin(
    sampled_gdf_within_mitte,
    bars_in_mitte_buffered,
    predicate='within'
)

# Step 15: Add bars and buffer zones to map
for _, row in bars_in_mitte.iterrows():
    folium.Circle(
        location=[row.geometry.centroid.y, row.geometry.centroid.x],
        radius=500,
        color='blue',
        fill=True,
        fill_opacity=0.2
    ).add_to(heatmap_map)

# Step 16: Save the map
heatmap_map.save('heatmap_with_mitte_bars_and_buffers.html')
print("Heatmap with Mitte, bars, and buffer zones saved as 'heatmap_with_mitte_bars_and_buffers.html'")

# Optional: Analysis of purchase activity within buffer zones
print("\nSummary of purchases within buffer zones:")
print(purchase_within_buffers.head())
print("\nTotal purchases within buffer zones:", len(purchase_within_buffers))
```
